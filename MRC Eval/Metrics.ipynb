```python
import pandas as pd
```

```python
df_test= pd.read_csv('dev_sal.csv')
```

```python
labels = list(df_test['label'])
```

```python
with open('bart_epoch3_labels.txt', 'r') as file:
    # Read all lines and store them in a list
    preds = file.readlines()

# Remove any trailing newline characters
preds = [line.strip() for line in preds]
```

```python
from sklearn.metrics import f1_score
def exact_match(list1, list2):
    return sum(a == b for a, b in zip(list1, list2)) / len(list1)

def f1_score_multiclass(true_labels, predictions):
    f1_micro = f1_score(true_labels, predictions, average='micro')
    f1_macro = f1_score(true_labels, predictions, average='macro')
    f1_weighted = f1_score(true_labels, predictions, average='weighted')
    
    return f1_micro, f1_macro, f1_weighted

classes = ["I", "E", "X", "N", "A", "#"]
for c in classes:
    truth = []
    p = []
    for i in range(0, len(labels)):
        if labels[i] == c:
            truth.append(1)
            if preds[i] == c:
                p.append(1)
            else:
                p.append(0)

    em_score = exact_match(truth, p)
    f1_micro, f1_macro, f1_weighted = f1_score_multiclass(truth, p)
    
    print("===============================================")

    print(f"{c} F1-Score (Micro): {f1_micro:.2f}")
    print(f"{c} F1-Score (Macro): {f1_macro:.2f}")
    print(f"{c} F1-Score (Weighted): {f1_weighted:.2f}")

    print(f"{c} Exact Match: {em_score * 100:.2f}%")
    
    print("===============================================\n")
```
