```python
import pandas as pd
final_df = pd.read_csv(f"BARTEPOCH3.csv")
import json
f = open("retrobartepoch3.json")
f = json.load(f)
retro_uq = []
retro_bart = []
for k, v in f.items():
    retro_uq.append(v[0])
    retro_bart.append(v[1])
final_df["OU retro answer span"] = retro_uq
final_df["SA retro answer span"] = retro_bart
f = open("epoch_2_predictions.json")
f = json.load(f)
sgnet_bart = []
for k, v in f.items():
    try:
        sgnet_bart.append(v[0])
    except:
        sgnet_bart.append(v)
f = open("epoch_2_nbest_predictions.json")
f = json.load(f)
sgnet_uq = []
for k, v in f.items():
    sgnet_uq.append(v[0]['text'])
final_df["OU sgnet answer span"] = sgnet_uq
final_df["SA sgnet answer span"] = sgnet_bart
new_data = pd.read_csv("dev_sal.csv")
context = new_data['context'].tolist()
unans = new_data["original_question"].tolist()
labels = new_data['label'].tolist()
final_df['context'] = context
final_df['uq'] = unans
final_df['labels'] = labels
with open("bart_epoch3.txt", 'r') as file:
    lines = file.readlines()

# Remove newline characters from each line
lines = [line.strip() for line in lines]
#lines.pop()
final_df['suggested aq'] = lines
with open("bart_epoch3_labels.txt", 'r') as file:
    lines = file.readlines()

# Remove newline characters from each line
lines = [line.strip() for line in lines]
final_df['predicted_labels'] = lines
#lines.pop()
```
