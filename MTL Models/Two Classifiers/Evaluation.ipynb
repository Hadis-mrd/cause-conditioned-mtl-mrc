```python
import pandas as pd
from tqdm import tqdm

# Define an empty list to hold the rows
rows = []

# Iterate through the DataFrame using iterrows
for index, row in tqdm(df_all.iterrows(), total=df_all.shape[0], desc="Processing rows"):
    if index % 100000 == 0: 
        print(index)

    qid = row['qid']
    context = row['context']
    OU_question = row['uq']
    label = row['labels']
    SA_question = row['suggested aq']

    model_data = [
        ('model 2', row['OU_unanswerability_model2'], row['SA_unanswerability_model2']),
        ('model 4', row['OU_unanswerability_model4'], row['SA_unanswerability_model4']),
        ('electra', row['OU_unanswerability_electra'], row['SA_unanswerability_electra']),
        ('roberta', row['OU_unanswerability_roberta'], row['SA_unanswerability_roberta']),
        ('sgnet', row['OU_unanswerability_sgnet'], row['SA_unanswerability_sgnet']),
        ('retro', row['OU_unanswerability_retro'], row['SA_unanswerability_retro'])]

    for model_name, OU_un, SA_un in model_data:
        row_data = [qid, context, OU_question, label, SA_question, OU_un , SA_un, model_name]
        rows.append(row_data)

# Convert the list of rows to a DataFrame
result_df = pd.DataFrame(rows, columns=['qid', 'context', 'uq', 'label', 'suggested aq', 
                                        'unanswerability on UQ question', 'unanswerability on SA question','model name'])
# result_df.to_csv('/home/hmoradis/UQA/Julien/Task21-Bart_no_Label/Evaluation/unanswerability_results_bart_no_label.csv')

result_df
```

```python
import pandas as pd

# Group by 'qid' and 'potential_UQ question'
grouped = result_df.groupby('qid')

# Initialize an empty list to hold the result
result_list = []
                                      
# Iterate over each group
for name, group in grouped:
    
    model_name_list = []
    sw = True
    qid = name  # unpack the group name
    
    # Assume that context, original_question, and original_answer_span are the same for each group
    context = group.iloc[0]['context']
    UQ_question = group.iloc[0]['uq']
    lebel = group.iloc[0]['label']
    SA_question = group.iloc[0]['suggested aq']
    
    
    # Iterate through each row in the group
    for index, row in group.iterrows():
        UQ_un = row['unanswerability on UQ question']
        SA_un = row['unanswerability on SA question']
        #### New eval (1,0,0) for model2,model4,electra,roberta AND (1,0) for retro and sgnet   
        model_n = row['model name']
        if UQ_un == 0 : ### unanswerable question in dev set should be labeled as an unanswerable question
            if SA_un == 1 :  ## suggested question should be answerable
                if row['model name'] not in model_name_list:
                    model_name_list.append(row['model name'])
    
    # Add the result to the list
    result_list.append({
        'qid': qid,
        'context': context,
        'UQ_question': UQ_question,
        'label': label,
        'SA_question': SA_question,
        'model_name_list': model_name_list,
        'len': len(model_name_list)
    })


# Create a new DataFrame from the result list
new_df = pd.DataFrame(result_list)
# new_df.to_csv('/home/hmoradis/UQA/Julien/Task21-Bart_no_Label/Evaluation/evalresult01-bart_no_label.csv')
# Display the new DataFrame
display(new_df)
```

```python
conditions = [
    ('len >= 1', new_df['len'] >= 1),
    ('len >= 2', new_df['len'] >= 2),
    ('len >= 3', new_df['len'] >= 3),
    ('len >= 4', new_df['len'] >= 4),
    ('len >= 5', new_df['len'] >= 5),
    ('len >= 6', new_df['len'] >= 6),
    ('len == 1', new_df['len'] == 1),
    ('len == 2', new_df['len'] == 2),
    ('len == 3', new_df['len'] == 3),
    ('len == 4', new_df['len'] == 4),
    ('len == 5', new_df['len'] == 5),
    ('len == 6', new_df['len'] == 6),
]

# Initialize an empty dictionary to hold the results
condition_counts = {}

# Iterate over each condition
for condition_name, condition in conditions:
    # Filter the DataFrame based on the current condition
    filtered_df = new_df[condition]
    
    # Count the number of records that meet the condition
    count = filtered_df.shape[0]
    
    # Store the count in the dictionary
    condition_counts[condition_name] = count

# Display the counts for each condition
for condition_name, count in condition_counts.items():
    print(f"Number of records where {condition_name}: {count}")
```
